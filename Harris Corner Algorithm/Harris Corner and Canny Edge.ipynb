# Harris Corner Algorithm

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('01.png')
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
img = np.float32(gray_image)
```

```python
corners = cv2.cornerHarris(img, 2, 5, 0.04)
```

```python
plt.subplot(1,1,1), plt.imshow(corners,cmap='RdGy')
plt.title('Harris Corner Detection'), plt.xticks([]), plt.yticks([])
```

```python
corners2 = cv2.dilate(corners,None, iterations = 3)
```

```python
img2 = cv2.imread('01.png')
```

```python
img2[corners2 > 0.01 * corners2.max()] = [255,0,0]
```

```python
plt.subplot(1,1,1), plt.imshow(img2,cmap='gray')
plt.title('Canny Edge Detection'), plt.xticks([]), plt.yticks([])
plt.show()
```
